## **Empresa e historia**

La empresa LengaMar Transporte y Turismo SpA es una compañía local ubicada en el sector costero de Caleta Lenga, en la Región del Biobío, dedicada al transporte de pasajeros y servicios turísticos en lancha dentro de la bahía de San Vicente y zonas cercanas.

La empresa fue fundada en 2018 por un grupo de emprendedores y pescadores de la zona que buscaban diversificar sus ingresos mediante el turismo costero y el transporte marítimo. Con el paso del tiempo, la empresa fue ampliando su flota de lanchas y comenzó a ofrecer distintos servicios como paseos turísticos, traslados a sectores cercanos de la costa y apoyo logístico para actividades recreativas y gastronómicas del sector.

Debido a que la operación de las lanchas depende fuertemente de las condiciones climáticas y marítimas, la empresa comenzó a registrar información diaria relacionada con el clima y las condiciones ambientales para mejorar la planificación de sus servicios.
