# 4.- Calcular e interpretar Insight

## - Insight 1 ¿En qué localidades y en qué meses se aprecia la mayor fluctuación de temperaturas (diferencia entre la temperatura máxima y mínima)?

```python
df["fecha"] = pd.to_datetime(df["date"]).dt.date

temperatura_dia = df.groupby(["Localidad","fecha"])["temperature_2m"].agg(['max', 'min']).reset_index().reset_index()
print(temperatura_dia.head(32))
```

```python
temperatura_dia['fluctuacion_diaria'] = temperatura_dia['max'] - temperatura_dia['min']
print(temperatura_dia.head(32))
```

```python
temperatura_dia['mes'] = pd.to_datetime(temperatura_dia['fecha']).dt.month

#print(temperatura_dia.head(32))

insight_1 = temperatura_dia.groupby(['Localidad', 'mes'])['fluctuacion_diaria'].mean().reset_index()

insight_1 = insight_1.sort_values(by='fluctuacion_diaria', ascending=False)
print(insight_1)
```

### **Respuesta**:
La mayor fluctuación de temperatura se observa en Temuco durante los meses de enero (1) y febrero (2). El promedio de fluctuación es de 14,8 °C en enero y 14,7 °C en febrero.

Esto significa que, en un día típico de verano, la temperatura en Temuco puede variar aproximadamente 15 °C entre el momento más frío y el más cálido del día.


## - Insight 2 Por localidad y mes, ¿cuál es la diferencia sistemática entre temperatura real y aparente?

```python
#temperatura_dia['sensacion_termica'] = df.apparent_temperature
#print(temperatura_dia.head(23))

temperatura_dia["dif_termica"] = df['apparent_temperature'] - df['temperature_2m']
#print(temperatura_dia.head(23))

insight_2 = temperatura_dia.groupby(['Localidad', 'mes'])['dif_termica'].mean().reset_index()
insight_2 = insight_2.sort_values(by='dif_termica', ascending=False)

print(insight_2)
```

### **Respuesta:**
Existe una diferencia sistemática negativa en gran parte del dataset. Esto significa que, de manera constante, la sensación térmica es menor que la temperatura real, por lo que los habitantes de estas localidades perciben más frío del que indica el termómetro.

La mayor diferencia negativa ocurre en Punta Arenas durante febrero (mes 2), con -1,73 °C. Aunque febrero corresponde al verano, los fuertes vientos característicos de la zona de Magallanes provocan que la sensación térmica disminuya casi 2 °C por debajo de la temperatura real.

# 5.- Identificar duplicados, atipicos y nulos

## - Valores nulos

```python
df.isna().sum().rename("Cantidad de nulos")
```

Gracias a la tabla, no se observan valores nulos en el conjunto de datos.

Sin embargo, si en el futuro se encontraran valores nulos, se debería evaluar cada caso para determinar la mejor forma de tratarlos. Algunas opciones posibles son:

**Imputación estadística (media, mediana o moda):**
Consiste en rellenar el valor faltante utilizando un estadístico calculado a partir del resto de la columna.

- Media: Ideal cuando los datos presentan una distribución relativamente simétrica.
- Mediana: Más recomendable cuando existen valores extremos (outliers) que pueden distorsionar el promedio.

**Imputación por contexto:**
Considerando que los datos corresponden a variables climáticas, es esperable que valores cercanos en el tiempo sean similares. Por ejemplo, la temperatura a las 03:00 AM probablemente sea cercana a la registrada a las 04:00 AM.

- Forward Fill (ffill): rellena el valor nulo utilizando el dato inmediatamente anterior.
- Interpolación: calcula un valor intermedio entre el dato anterior y el siguiente, lo que suele ser una de las técnicas más precisas para datos climáticos.

**Eliminación de filas:**
Esta es la opción más extrema. Se utiliza cuando la cantidad de valores nulos es muy baja (por ejemplo, menos del 1% o 5%) y su eliminación no afecta la representatividad del conjunto de datos.
En cambio, si una variable presenta un porcentaje muy alto de valores nulos (por ejemplo, más del 50%), podría ser recomendable no utilizar esa variable en los modelos de análisis o de machine learning.

## - Valores atipicos

```python
seleccion = ["temperature_2m"]
# Genera, por separado, boxplot para cada variable numérica del conjunto indicado y luego integra en una sola gráfica
plt.figure(figsize=(12,6))

for i, columna in enumerate(seleccion, 1):
    plt.subplot(1, len(seleccion), i)
    sb.boxplot(y=df[columna])
    plt.title(columna, fontsize=14, fontweight="bold")

plt.tight_layout()
plt.show()
```

- El borde inferior de la caja corresponde a 7,05 °C (25%).
- El borde superior de la caja corresponde a 13,7 °C (75%).

**Conclusión:**
La mayoría de las temperaturas registradas en estas cuatro localidades chilenas se concentran entre 7 °C y 14 °C, lo que indica un predominio de temperaturas frescas dentro del conjunto de datos.

Los bigotes o líneas negras que se extienden desde la caja representan los valores que aún se consideran dentro del rango normal de la distribución.
Cualquier dato que se encuentre más allá de estos límites se considera un valor atípico (outlier), ya que se aleja significativamente del comportamiento general de los datos.

**¿Qué hacer si existieran inconsistencias?**

Si se hubieran detectado inconsistencias en los datos, se podrían considerar las siguientes acciones:

- **Sustitución lógica:**
Si el valor es levemente negativo (por ejemplo, -0,1), se podría corregir a 0, asumiendo que se trata de un error de medición o de calibración del instrumento.

- **Imputación por vecinos (interpolación):**
Debido a que las variables climáticas suelen cambiar de forma gradual, si faltara un dato a las 10:00, se podría estimar utilizando el promedio entre los valores registrados a las 09:00 y 11:00.

- **Eliminación:**
Si el valor es completamente irreal o corresponde a un outlier técnico, se podría eliminar la fila o marcar el valor como nulo, evitando que afecte negativamente los cálculos o el entrenamiento de modelos de machine learning.
